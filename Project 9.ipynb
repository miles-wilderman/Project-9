For this we will be focusing on the 24 hour new snowfall column. 

```python
#Preprocessing
JacksonHole['Date'] = pd.to_datetime(JacksonHole['Date'])
JacksonHole['month'] = JacksonHole['Date'].dt.strftime('%B')

Snowbird['Date'] = pd.to_datetime(Snowbird['Date'])
Snowbird['month'] = Snowbird['Date'].dt.strftime('%B')

Telluride['Date'] = pd.to_datetime(Telluride['Date'])
Telluride['month'] = Telluride['Date'].dt.strftime('%B')

Whistler['Date'] = pd.to_datetime(Whistler['Date'])
Whistler['month'] = Whistler['Date'].dt.strftime('%B')
```

```python
JacksonHole['24 hr New Snow'] = JacksonHole['24 hr New Snow'].str.replace('cm', '')
JacksonHole['24 hr New Snow'] = JacksonHole['24 hr New Snow'].astype(int)

Snowbird['24 hr New Snow'] = Snowbird['24 hr New Snow'].str.replace('cm', '')
Snowbird['24 hr New Snow'] = Snowbird['24 hr New Snow'].astype(int)

Telluride['24 hr New Snow'] = Telluride['24 hr New Snow'].str.replace('cm', '')
Telluride['24 hr New Snow'] = Telluride['24 hr New Snow'].astype(int)

Whistler['24 hr New Snow'] = Whistler['24 hr New Snow'].str.replace('cm', '')
Whistler['24 hr New Snow'] = Whistler['24 hr New Snow'].astype(int)
```

```python
JacksonHole['Base Depth'] = JacksonHole['Base Depth'].replace('cm', '')
JacksonHole['Base Depth'] = JacksonHole['Base Depth'].astype(int)

Snowbird['Base Depth'] = Snowbird['Base Depth'].replace('cm', '')
Snowbird['Base Depth'] = Snowbird['Base Depth'].astype(int)

Telluride['Base Depth'] = Telluride['Base Depth'].replace('cm', '')
Telluride['Base Depth'] = Telluride['Base Depth'].astype(int)

Whistler['Base Depth'] = Whistler['Base Depth'].replace('cm', '')
Whistler['Base Depth'] = Whistler['Base Depth'].astype(int)
```

```python
JacksonHole
```

```python
merged_df = pd.concat([JacksonHole, Snowbird, Telluride, Whistler], axis=0, ignore_index=True)
```

```python
merged_df
```

```python
X = merged_df.iloc[:, [1,2,4]].values
y = merged_df.iloc[:, 3].values
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder_X = LabelEncoder()
X[:, 2] = labelEncoder_X.fit_transform(X[:, 2])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scale_X = StandardScaler()
X_train = scale_X.fit_transform(X_train)
X_test = scale_X.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression( random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.linear_model import Ridge

model = Ridge(alpha=0.1)

model.fit(X_train, y_train)
```

```python
y_pred2 = model.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.mean_absolute_error(y_test, y_pred2))
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor()

gb_model.fit(X_train, y_train)
```

```python
y_pred3 = gb_model.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.mean_absolute_error(y_test, y_pred3))
```

The error is down a lot but it is still a very bad model. This is most likely due to just a lack of data. My next step would be to combine temperature and weather data with this snow data to get more accurate results.
